# WeatherPy
----


1) As a whole, Max Temperatures by Cities tend to increase the closer to the equator at 0 degrees latitude, and decrease when moving further away from the equator, regardless of direction (North or South).

2) Within cities in the Southern Hemisphere, there is a small negative correlation between Latitude and Humidity, in that as Latitude moves closer to the equator (zero degrees), Humidity begans to decrease.

3) within cities in the southern Hemisphere, there is a small negative correlation between cloudiness and latitude, in that as latitude moves closer to the equator (zero degrees), cloudiness begans to slowly decrease.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
import warnings
warnings.filterwarnings('ignore')

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# API sample check for index reference
# https://api.openweathermap.org/data/2.5/weather?appid=b5d3842347223a2380a7fdfd5ff72b29&units=metric&q=rikitea
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial" # try 'metric' for units if imperial gives any problems

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="

# set up lists to hold city response info
City_Data = []
record = 0
set_num = 1

# print start of API Call Series
print("Beginning Data Retrieval\n")
print("--------------------------------\n")

# Loop through the cities list and perform a request for data on each
for city in cities:
    # if city is found in API...
    try:
        response = requests.get(query_url + city).json() # json request
        city_lat = response['coord']["lat"] # latitude
        city_lng = response['coord']['lon'] # longitude
        city_maxTemp = response['main']['temp_max'] # max temperature
        city_humidity = response['main']['humidity'] # humidity
        city_cloud = response['clouds']['all'] # clouds
        city_windspeed = response['wind']['speed'] # wind speed
        city_country = response['sys']['country'] # country abbrev.
        city_dt = response['dt'] # date
        
        # print city if data found
        print(f"Processing Record {record} of Set {set_num} | {city}")
        
        # start counter of records counter for each city in each set found in API
        record += 1
        
        # once record 50 is printed for each set, record number will reset back to zero and set number will increase by one
        if record % 51 == 0:
            record = 0
            set_num += 1
            
        # continue record count with current set number if not 50th record of set printed   
        else:
            pass

        # append city data into City Data list using dictionary
        City_Data.append({
            'City':city,
            'Lat':city_lat,
            'Lng':city_lng,
            'maxTemp':city_maxTemp,
            'humidity':city_humidity,
            'clouds':city_cloud,
            'windspeed':city_windspeed,
            'country':city_country,
            'date':city_dt
        })
        
    # if city not found in API...    
    except:
        print('City not found. Skipping...')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create Data Frame
City_Data_Frame = pd.DataFrame(City_Data)

# Display Data Frame
City_Data_Frame.head()

# Export data frame to csv
City_Data_Frame.to_csv("Documents/City_Data_Frame.csv")
```

## Linear Regression

```python
#  Get the indices of cities in the northern hempishere and create data frame
North_Hemisphere = clean_city_data[(clean_city_data['Lat'] >= 0)]

# create Data Frame for Northern Hemisphere
NorthSphereDF = pd.DataFrame(North_Hemisphere)

#  Get the indices of cities in the southern hempishere and create data frame
South_Hemisphere = clean_city_data[(clean_city_data['Lat'] < 0)]

# create Data Frame for Southern Hemisphere
SouthSphereDF = pd.DataFrame(South_Hemisphere)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Add the linear regression equation and line to plot
x_values = NorthSphereDF['Lat']
y_values = NorthSphereDF['maxTemp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Print the r-value:
print(f"The r-value is: {rvalue**2}")

# Create Scatter Plot comparing City Latitude and Max Temperature using matplotlib
plt.scatter(x_values,y_values)
plt.xlabel('Latitude')
plt.ylabel('Max Temperature')
plt.plot(x_values,regress_values,'r-')
plt.annotate(line_eq,(x_values[0::-1],y_values[0::-1]),fontsize=15,color='red')
plt.savefig('Documents/Northern Hemisphere - Max Temp vs. Latitude Linear Regression.png')
plt.show()
```
